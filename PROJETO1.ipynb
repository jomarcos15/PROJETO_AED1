```python
import random

def jogo_adivinhacao_completo():
    # 1. Escolhe um n√∫mero aleat√≥rio entre 1 e 100
    numero_secreto = random.randint(1, 100)
    tentativas = 0
    palpite = 0
    # Inicializa a lista para armazenar o hist√≥rico de palpites
    historico_palpites = []

    print("Bem-vindo ao Jogo de Adivinha√ß√£o!")
    print("Estou pensando em um n√∫mero entre 1 e 100.")
    
    # Loop principal do jogo
    while palpite != numero_secreto:
        try:
            # Pede o palpite do usu√°rio
            palpite = int(input("\nDigite seu palpite: "))
            
            # Adiciona o palpite √† lista antes de verificar se est√° correto
            historico_palpites.append(palpite) 
            tentativas += 1 
            
            # Fornece feedback (maior ou menor)
            if palpite < numero_secreto:
                print("Seu palpite √© **menor** que o n√∫mero secreto. Tente novamente!")
            elif palpite > numero_secreto:
                print("Seu palpite √© **maior** que o n√∫mero secreto. Tente novamente!")
            
        except ValueError:
            # Lida com entradas n√£o num√©ricas
            print("Entrada inv√°lida. Por favor, digite um n√∫mero inteiro.")
            
    # Mensagem de sucesso quando o usu√°rio acerta
    print(f"\nüéâ Parab√©ns! Voc√™ acertou o n√∫mero secreto: {numero_secreto}!")
    print(f"Voc√™ precisou de {tentativas} tentativas.")

    # Exibe o hist√≥rico de palpites
    print("\n--- Hist√≥rico de Palpites ---")
    print(f"Seus palpites foram: {historico_palpites}")
    print("----------------------------")

# Inicia o jogo
jogo_adivinhacao_completo()
```
